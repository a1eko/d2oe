```python
import os

network_path = os.path.join("networks", "d2oe0")
config_file = os.path.join("config", "network-d2oe0.json")
```

## Create the mesh for the neurons

```python
n_total = 10000
mesh_file = f"config/mesh/mesh-{n_total}.obj"
if not os.path.isfile(mesh_file):
    from snudda.place.create_cube_mesh import create_cube_mesh
    density = 80500
    side_len = (n_total/density)**(1/3)*1e-3
    create_cube_mesh(file_name=mesh_file, centre_point=(0,0,0), side_len=side_len)
    print(f"Created {mesh_file}")
else:
    print(f"Mesh file already exists")
```

## Generate network

```python
from snudda import Snudda

snd = Snudda(network_path)
snd.import_config(network_config_file=config_file, overwrite=True)
```

```python
snd.create_network()
```

## Analyse connectivity

```python
from snudda.analyse.analyse_striatum import SnuddaAnalyseStriatum

nas = SnuddaAnalyseStriatum(network_path, volume_type="cube")

y_max_H = None  # 0.5
dist3D = True

nas.plot_connection_probability("dSPN", "iSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[50e-6, 100e-6],
                                exp_data=[3 / 47.0, 3 / 66.0],
                                exp_data_detailed=[(3, 47), (3, 66)],
                                y_max=y_max_H)

nas.plot_connection_probability("dSPN", "dSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[50e-6, 100e-6],
                                exp_data=[5 / 19.0, 3 / 43.0],
                                exp_data_detailed=[(5, 19), (3, 43)],
                                y_max=y_max_H)

nas.plot_connection_probability("iSPN", "dSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[50e-6, 100e-6],
                                exp_data=[13 / 47.0, 10 / 80.0],
                                exp_data_detailed=[(13, 47), (10, 80)],
                                y_max=y_max_H)

nas.plot_connection_probability("iSPN", "iSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[50e-6, 100e-6],
                                exp_data=[14 / 39.0, 7 / 31.0],
                                exp_data_detailed=[(14, 39), (7, 31)],
                                y_max=y_max_H)

nas.plot_connection_probability("FS", "iSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[100e-6, 150e-6, 250e-6],
                                exp_data=[6 / 9.0, 21 / 54.0, 27 / 77.0],
                                exp_data_detailed=[(6, 9), (21, 54), (27, 77)],
                                y_max=None)

nas.plot_connection_probability("FS", "dSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[100e-6, 150e-6, 250e-6],
                                exp_data=[8 / 9.0, 29 / 48.0, 48 / 90.0],
                                exp_data_detailed=[(8, 9), (29, 48), (48, 90)],
                                y_max=None)

nas.plot_connection_probability("FS", "FS",
                                dist_3d=dist3D,
                                exp_max_dist=[250e-6],
                                exp_data=[7 / 12.0],
                                exp_data_detailed=[(7, 12)])

nas.plot_connection_probability("FS", "FS",
                                dist_3d=dist3D,
                                connection_type="gapjunctions",
                                exp_max_dist=[200e-6, 250e-6, 250e-6],
                                exp_data=[6/78.0, 2 / 6.0, 3 / 7.0],
                                exp_data_detailed=[(6, 78), (2, 6), (3, 7)])

nas.plot_connection_probability("LTS", "dSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[250e-6],
                                exp_data=[2 / 60.0],
                                exp_data_detailed=[(2, 60)],
                                x_max=500)

nas.plot_connection_probability("LTS", "iSPN",
                                dist_3d=dist3D,
                                exp_max_dist=[250e-6],
                                exp_data=[2 / 60.0],
                                exp_data_detailed=[(2, 60)],
                                x_max=500)

nas.plot_connection_probability("FS", "LTS",
                                dist_3d=dist3D,
                                exp_max_dist=[250e-6],
                                exp_data=[2.0 / 12],
                                exp_data_detailed=[(2, 12)])

```
